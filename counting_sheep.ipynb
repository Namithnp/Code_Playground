```python
import time
import random
import numpy as np


class SleepSimulator:
    def __init__(self, dream_list):
        self.dreams = dream_list
        self.comf_lvl = np.random.randint(0, 10)
        bins = [5, 8]   #Different segments of comfort levels between 0 to 10
        self.segs = np.digitize(self.comf_lvl, bins)
        #print(segs)
        np.random.seed(self.comf_lvl)


    def countdown(self):
        countdown_segs = {
            0 : (31, 50),
            1 : (21, 30),
            2 : (0, 20)
        }
        low, high = countdown_segs.get(self.segs)
        sleep_countdown = random.randint(low, high)
        #print(sleep_countdown)
        print("Counting sheep.....")
        for i in range(sleep_countdown, 10, -1):
            print(i)
            time.sleep(1)
        slow_ctdn_range = np.arange(1.5, 6.5, 0.5) #Stretched out countdown after 10
        sleepy_time = [i for i in range(10, 0, -1)]
        slow_ctdn = dict(zip(sleepy_time, slow_ctdn_range))
        for t in slow_ctdn.keys():
            print(f'{t}...')
            time.sleep(slow_ctdn[t])

    def dream(self):
        if self.segs == 2:
            dream_num = random.choice(self.dreams)
            print(f'You are dreaming of {dream_num}!')

dream_list = ["dream1", "dream2", "dream3", "dream4", "dream5"]

simulator = SleepSimulator(dream_list)
simulator.countdown()
simulator.dream()
```
